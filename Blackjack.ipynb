```python
class Card():
    
    def __init__(self,suit,rank):
        self.suit = suit
        self.rank = rank
    
    def __str__(self):
        return str(f'{self.rank} of {self.suit}')
```

```python
class Deck():
    
    def __init__(self):
        self.deck = []  # start with an empty list
        for suit in suits:
            for rank in ranks:
                self.deck.append(Card(suit,rank))
                
    
    def __str__(self):
        comp_deck = ''
        for card in self.deck:
            comp_deck += '\n'+card.__str__()
        return 'The deck contains: '+comp_deck
        

    def shuffle(self):
        random.shuffle(self.deck)
        
    def deal(self):
        single_card = self.deck.pop()
        return single_card
```

```python
class Hand:
    def __init__(self):
        self.cards = []  # start with an empty list as we did in the Deck class
        self.value = 0   # start with zero value
        self.aces = 0    # add an attribute to keep track of aces
    
    def add_card(self,card):
        self.cards.append(card)
        self.value += values[card.rank]
        if 'Ace' in self.cards:
            self.aces += 1
    
    def adjust_for_ace(self):
        while self.value > 21 and self.aces:
            self.value -= 10
            self.aces -= 1
```

```python
class Chips():
    
    def __init__(self,total=100):
        self.total = total  # This can be set to a default value or supplied by a user input
        self.bet = 0
        
    def win_bet(self):
        self.total += self.bet
    
    def lose_bet(self):
        self.total -= self.bet
```

```python
def take_bet(chips):
    
    while True:
    
        try:
            chips.bet = int(input('How much would you like to bet? '))

        except ValueError:
            print('Sorry please provide a number of chips')
        else:
            if chips.bet > chips.total:
                print(f'Sorry, you dont have enough chips! You have {chips.total} chip to bet with!')
            else:
                break
        
```

```python
def hit(deck,hand):
    
    hand.add_card(deck.deal())
    hand.adjust_for_ace()   
```

```python
def hit_or_stand(deck,hand):
    global playing  # to control an upcoming while loop
    
    while True:
        x = input('Hit or Stand? enter h or s')
        
        if x[0].lower() == 'h':
            hit(deck,hand)
            
        elif x[0].lower() == 's':
            print('Player Stands. Dealers turn.')
            playing = False
            
        else:
            print("I'm sorry. I didn't get that. Please provide h or s...")
            
        break
```
